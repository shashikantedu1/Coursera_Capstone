## 3. Methodology Section

Clustering and Segmentation Approach:

Localities comparision on the basis of available venues in the neighborhood will be of very much help to wide variety of audience. 

Approach was to form clusters in Toronto city on the basis of neighborhood, most frequent searched or accessed venues. Segmented localities on the basis of available ameneties will help people to identify and take decisions to meet their requirements.

In order to do that, we need to cluster data which is a form of unsupervised machine learning: k-means clustering algorithm.

In the above image, neighborhoods have been classfied into different cluster on the basis of similarities of the most common venues available in the localities.
